```python
#Identifying unique municipality codes
unique_cumun = df['CUMUN'].drop_duplicates()
```

```python
unique_cumun
```

```python
#Defining blank dataframe to collect merged polygon data
df_cumun_polygons = pd.DataFrame()
```

```python
#Merging census sections into municipalities
for icumun in unique_cumun:
    polygons_to_merge = df[df.CUMUN==icumun]
    merged_polygon = polygons_to_merge["geometry"].unary_union
    temp_df = pd.DataFrame({"CUMUN":[icumun],"geometry_combined":[merged_polygon]})
    print(icumun)
    df_cumun_polygons = pd.concat([df_cumun_polygons, temp_df], ignore_index=True)

    
```

```python
#Saving the municipality file. THis file it too big to upload to github
gdf_municipio_polygons = gp.GeoDataFrame(df_cumun_polygons,crs='epsg:4326',geometry="geometry_combined")
gdf_municipio_polygons.to_file('municipio_polygons.geojson',driver = 'GeoJSON')
```

```python
#deleting municipio_polygons.geojson because it's too large to upload in git hub
#Reducing the size of the file by smoothening out the edges. Geopandas has inbuilt functions to do this
# df4 = gp.read_file("municipio_polygons.geojson")
df4 = gdf_municipio_polygons
```

```python
#Creating smoothened polygons
tolerance = 0.001  # Adjust the tolerance value as needed
df4['geometry'] = df4['geometry'].simplify(tolerance)
```

```python
#Saving smoothened file
output_file = 'simplified_municipio_polygons.geojson'
df4.to_file(output_file, driver='GeoJSON')
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Polygon
```

```python
# Load the district data
districts = gpd.read_file('madrid_distrito_polygons.geojson')
```

```python
# Load the postcode data
postcodes = gpd.read_file('Carto_postal_codes.geojson')
```

```python
# Filter postcodes for Madrid
postcodes = postcodes[postcodes['COD_POSTAL'].str.startswith('280')]
```

```python
# Create a list to store all the overlap data
overlap_data = []
```

```python
districts
```

```python
postcodes
```

```python
# Iterate through each district
for index, district in districts.iterrows():
    # Store top six postcodes and their percentage overlaps
    top_postcodes = ['']*6
    top_percentages = [0]*6
    
    # Get the area of the current district region
    district_area = district['geometry'].area
    
    # Iterate through each postcode
    for _, postcode in postcodes.iterrows():
        # Get the intersection of the current district and postcode region
        intersection = district['geometry'].intersection(postcode['geometry'])
        
        # Calculate the overlap percentage
        overlap_percentage = intersection.area / district_area * 100
        
        # Insert the postal code and its overlap percentage into the sorted lists
        for i in range(6):
            if overlap_percentage > top_percentages[i]:
                top_percentages.insert(i, overlap_percentage)
                top_postcodes.insert(i, postcode['COD_POSTAL'])
                # Keep only the top six
                top_percentages = top_percentages[:6]
                top_postcodes = top_postcodes[:6]
                break

    # Add the top six postcodes and their overlap percentages to the overlap data
    for i in range(6):
        overlap_data.append({
            'CUDIS': district['CUDIS'],
            'COD_POSTAL': top_postcodes[i],
            'Connection_order': i+1,
            'Percentage_overlap': top_percentages[i]
        })
```

```python
# Convert the overlap data to a DataFrame
overlap_df = pd.DataFrame(overlap_data)
```

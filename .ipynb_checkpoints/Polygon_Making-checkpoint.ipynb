```python
import geopandas as gp
import fiona
import matplotlib
import pandas as pd
```

```python
df = gp.read_file("SECC_CPV_E_20111101_01_R_INE (1).json")
```

```python
df
```

```python
df2 = df.query('NMUN =="Madrid"')
```

```python
df2
```

```python
polygons_to_merge = df.query('NMUN =="Madrid" and CDIS == "02"')
polygons_to_merge
```

```python
merged_polygon = polygons_to_merge['geometry'].unary_union

merged_polygon
```

```python
merged_polygon_df = pd.DataFrame({'CDIS':['02'], 'geometry':[merged_polygon]})
```

```python
merged_polygon_df
```

```python
gdf_individual_polygons = gp.GeoDataFrame(polygons_to_merge,crs='epsg:4326')
```

```python
gdf_merged_polygon = gp.GeoDataFrame(merged_polygon_df,crs='epsg:4326')
```

```python
gdf_individual_polygons.to_file('individual_polygons.geojson', driver ='GeoJSON')
```

```python
gdf_merged_polygon.to_file('merged_polygon.geojson',driver = 'GeoJSON')
```
